```python
df=pd.read_excel('C:\\FreqRelation_Audit\\Data.xlsx')
df1=df.LRFChannelPlan.str.replace(' ',',')
df2=df.SiteEarfcn.str.replace(' ',',')
df3=df.CellID
df4 = pd.concat([df1, df2, df3], axis= 1)
def find_arfcn(row):
    x=row['SiteEarfcn'].split(' ')
    y=row['LRFChannelPlan'].split(' ')
    z=[i in y for i in x]
    return all(z)
df4['Search']=df.apply(find_arfcn,axis=1)



inter = df4.SiteEarfcn.str.split(',').apply(set) - df4.LRFChannelPlan.str.split(',').apply(set).values
df4['EarfcnToDelete'] = inter.str.join(',')

inter = df4.LRFChannelPlan.str.split(',').apply(set) - df4.SiteEarfcn.str.split(',').apply(set).values
df4['EarfcnToAdd'] = inter.str.join(',')


df4.to_excel('C:\\FreqRelation_Audit\\Output_File.xlsx')
```
