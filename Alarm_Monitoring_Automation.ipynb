```python
import os
import pandas as pd
import string 
import openpyxl
import numpy as np
from openpyxl import load_workbook
from pandas import DataFrame, merge,read_excel
cwd=r'C:\\Alarm_monitoring'
cwd=cwd+'\\Nokia'
file_list=os.listdir(cwd)
csv_files=[cwd+'\\'+i for i in file_list if i.find('csv')>-1]
xls_files=[cwd+'\\'+i for i in file_list if i.find('xls')>-1]
df_list=[]

for i in csv_files:
    temp_df=pd.read_csv(i)
    df_list.append(temp_df)
df1_list=[]

for j in xls_files:
    temp1_df=pd.read_excel(j)
    df1_list.append(temp1_df)
    
df_SAMS = pd.DataFrame.from_dict (df_list[0])
df_macro = pd.DataFrame.from_dict (df_list[1])
df_mm = pd.DataFrame.from_dict (df_list[2])
df_SITE = pd.DataFrame.from_dict (df1_list[0])
df_mmSITE = pd.DataFrame.from_dict (df1_list[1])

df_macro[(df_macro['Severity'] !='MINOR')]
df_mm[(df_mm['Severity'] !='MINOR')]


df_macro['MRBTS'] = df_macro['Distinguished Name'].str[10:21]
df_mm['MRBTS'] = df_macro['Distinguished Name'].str[10:21]


df2=df_SITE[['MRBTS', 'Cascade']]
df11=df_SAMS[['SAMS#', 'Status', 'Cascade']]
df12=df_mmSITE[['MRBTS', 'Cascade']]
df13=pd.merge(df_macro, df2, left_index=True, right_index=True)
df15=pd.merge(df13, df11, left_index=True, right_index=True)
df16=pd.merge(df12, df_mm, left_index=True, right_index=True)
df17=pd.merge(df16, df11, on='Cascade', how='left')
df15.to_excel('C:\\Alarm_monitoring\\Nokia\\Alm_Nokmacro.xlsx', sheet_name='Sheet1')
df17.to_excel('C:\\Alarm_monitoring\\Nokia\\Alm_Nokmm.xlsx', sheet_name='Sheet1')

cwd=r'C:\\Alarm_monitoring'
cwd=cwd+'\\Ericsson'
file_list=os.listdir(cwd)
csv_files=[cwd+'\\'+i for i in file_list if i.find('csv')>-1]
xls_files=[cwd+'\\'+i for i in file_list if i.find('xls')>-1]
df4=pd.DataFrame()

df_list=[]

for i in csv_files:
    temp_df=pd.read_csv(i)
    df_list.append(temp_df) 


df1_list=[]

for j in xls_files:
    temp1_df=pd.read_excel(j)
    df1_list.append(temp1_df)



dfA = pd.DataFrame.from_dict (df_list[1])
df_LD = pd.DataFrame.from_dict (df1_list[0])
df_SITE = pd.DataFrame.from_dict (df1_list[1])

dfA.to_excel('C:\\Alarm_monitoring\\Ericsson\\Book1.xlsx', sheet_name='Sheet')
df_FM=pd.read_excel('C:\\Alarm_monitoring\\Ericsson\\Book1.xlsx', sheet_name='Sheet')

df_FM=df_FM[(df_FM.Severity!='MINOR') &  (df_FM.Severity!='WARNING')]

df_FM['Network Element'].str.split('=') 
df_FM['eNB']=df_FM['Network Element'].apply(lambda x:x.split("=")[-1])

df7=df_LD[['eNB','Todays Remarks']]
df8=df_SITE[['eNB','Cascade']]
df9=df_SAMS[['Cascade','SAMS#','Status']]
df9.rename(columns = {'Cascade' : 'Cascade_y'}, inplace=True)
df3=pd.merge(df8, df_FM, left_index=True, right_index=True)
df4=pd.merge(df7, df3, left_index=True, right_index=True)
df5=pd.merge(df9, df4, left_index=True, right_index=True)
df5.rename(columns = {'eNB_x' : 'eNB'}, inplace=True)
del df5['eNB_y']
del df5['eNB']
df5 = df5[(df5['Specific Problem'] != 'Fan Failure') & (df5['Specific Problem'] != 'Battery Disconnected') & (df5['Specific Problem'] != 'Battery Missing') & (df5['Specific Problem'] != 'Battery Failure Battery Disconnected')]
df5 = df5[(df5['Specific Problem'] != 'Circuit Breaker Tripped') & (df5['Specific Problem'] != 'Climate Inconsistent Configuration') & (df5['Specific Problem'] != 'External Link Failure') & (df5['Specific Problem'] != 'External Alarm')]
df5 = df5[(df5['Specific Problem'] != 'SW Error') & (df5['Specific Problem'] != 'Power Loss') & (df5['Specific Problem'] != 'Temperature Sensor Failure') & (df5['Specific Problem'] != 'Enclosure Smoke Detector Failure') & (df5['Specific Problem'] != 'Loss of Mains')]
df5 = df5[(df5['Specific Problem'] != 'Fan Speed Continuously High') & (df5['Specific Problem'] != 'Operating Temperature Too High Battery Disconnected') & (df5['Specific Problem'] != 'Enclosure Door Open')]
df5.to_excel('C:\\Alarm_monitoring\\Ericsson\\Alarm_file.xlsx', sheet_name='Sheet1')
```
